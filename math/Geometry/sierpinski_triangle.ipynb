```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000

sx = np.zeros(N)
sy = np.zeros(N)
```

```python
for i in range(1,N):
    j = np.random.randint(1,4)

    sx[i] = sx[i-1]/2 + j-1
    sy[i] = sy[i-1]/2
    if j == 2:
        sy[i] += 2
```

```python
plt.plot(sx, sy, 'b.', markersize = 1)
plt.title('Sierpinski Triangle')
plt.axis('off')
plt.show()
```
